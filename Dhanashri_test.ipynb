```python
bus = smbus.SMBus(1)
device_address = 0x48

def read_temperature():
    temp = bus.read_byte_data(device_address, 0x00)
    if -40 <= temp <= 125:
        print("Temperature PASS")
    else:
        print("Temperature FAIL")

def set_threshold(low, high):
    bus.write_byte_data(device_address, 0x02, low)
    bus.write_byte_data(device_address, 0x03, high)

read_temperature()
set_threshold(10, 50)
```

```python
# looping through a dictionary
fav_name = {'eran': 1, 'omkar': 2, 'dhanashri': 3}
for name, number in fav_name.items():
    print(f"{name} loves number {number}.") 

# looping through a dictionary keys
fav_name = {'eran': 1, 'omkar': 2, 'dhanashri': 3}
for name in fav_name.keys():
    print(name)

#looping through a value in a dictionary
fav_name = {'eran': 1, 'omkar': 2, 'dhanashri': 3}
for number in fav_name.values():
    print(number)   
```

```python
# abstarction example
from abc import ABC, abstractmethod
class Vehicle(ABC):
    
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    
    def start(self):
        print("Car starts")

c = Car()
c.start()
```

```python
# reverse a list
def reverse_list(my_list):
    return my_list[::-1]


numbers = [1, 2, 3, 4, 5]
reversed_numbers = reverse_list(numbers)

print(reversed_numbers)
```

```python
class ListOperations:

    def reverse_list(self, my_list):
        return my_list[::-1]

# Create object
obj = ListOperations()

numbers = [1, 2, 3, 4, 5]

result = obj.reverse_list(numbers)

print("Original List:", numbers)
print("Reversed List:", result)
```

```python
# sorting a list
class ListOperations:

    def sort_list(self, my_list):
        return sorted(my_list)

# Create object
obj = ListOperations()

numbers = [5, 2, 9, 1, 3]

result = obj.sort_list(numbers)

print("Original List:", numbers)
print("Sorted List:", result)
```
