```python
from smbus2 import SMBus

bus = SMBus(1)
device_address = 0x48

def read_temperature():
    temp = bus.read_byte_data(device_address, 0x00)
    if -40 <= temp <= 125:
        print("Temperature PASS")
    else:
        print("Temperature FAIL")

def set_threshold(low, high):
    bus.write_byte_data(device_address, 0x02, low)
    bus.write_byte_data(device_address, 0x03, high)

read_temperature()
set_threshold(10, 50)
```

```python
class Fraction:
    def __init__(self, x, y):
        self.num = x
        self.den = y

    def __str__(self):
        return '{}/{}'.format(self.num, self.den)
    
    def __add__(self, other):
        new_num = self.num * other.den + other.num * self.den
        new_den = self.den * other.den
        return '{}/{}'.format(new_num, new_den)
    
fr1 = Fraction(1, 2)
fr2 = Fraction(1, 3)
print(fr1)
print(fr2)
print(fr1 + fr2)    
```

```python
def factorial(n):
    if n == 0:          # Base case
        return 1
    else:               # Recursive case
        return n * factorial(n - 1)

print(factorial(5))
```

```python
# sorting a list without using class

def sort_list(lst):
    return sorted(lst)

print(sort_list([5, 2, 9, 1]))
```

```python
# sorting a list using class and method Encapsulation

class ListOperations:
    
    def __init__(self, lst):
        self.lst = lst   # data

    def sort_list(self):
        return sorted(self.lst)   # method

obj = ListOperations([5, 2, 9, 1])
print(obj.sort_list())
```

```python
# sorting a list using class and method Abstraction

class ListOperations:
    
    def __init__(self, lst):
        self.lst = lst

    def sort_list(self):
        return sorted(self.lst)   # internal logic hidden

obj = ListOperations([5, 2, 9, 1])
print(obj.sort_list())

```

```python
# sorting a list using class and method with Inheritance

class ListOperations:
    
    def __init__(self, lst):
        self.lst = lst

class AdvancedListOperations(ListOperations):  # Inheriting
    
    def sort_list(self):
        return sorted(self.lst)

obj = AdvancedListOperations([5, 2, 9, 1])
print(obj.sort_list())

```

```python
# sorting a list using class and method with Polymorphism

class ListOperations:
    
    def sort_list(self):
        print("Normal sorting")

class ReverseSort(ListOperations):
    
    def sort_list(self):
        print("Reverse sorting")

obj1 = ListOperations()
obj2 = ReverseSort()

obj1.sort_list()   # Normal sorting
obj2.sort_list()   # Reverse sorting
```
