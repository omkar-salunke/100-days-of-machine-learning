```python
def fib(n):
    if n <=0:
        return 0
    elif n==1:
        return 1
    else:
        return fib(n-1)+fib(n-2)

fib(38)
```

```python
def fib(n):
    st_data = {}
    if n <=0:
        st_data[0] = 0
        return 0
    elif n==1:
        st_data[1] = 0
        return 1
    else:
        if n-1 in st_data:
            a = st_data[n-1]
        else:
            a = fib(n-1)
        if n-2 in st_data:
            b = st_data[n-2]
        else:
            b = fib(n-2)
        st_data[n] = a+b
        return st_data[n]

fib(38)
```

```python
def fib(n,d):

    if n in d:
        return d[n]
    else:
        d[n] = fib(n-1,d)+fib(n-2,d)
        return d[n]
d = {0:0,1:1}

fib(38,d)
```
